```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
```

```python
column_names = data.columns
df = pd.DataFrame(columns=column_names)
df = pd.concat([df,data],axis=0)
```

```python
df
```

```python
df[["Income","Months Since Policy Inception","Months Since Policy Inception","Number of Open Complaints","Number of Policies"]] = df[["Income","Months Since Policy Inception","Months Since Policy Inception","Number of Open Complaints","Number of Policies"]].apply(pd.to_numeric)
```

```python
df
```

```python
df_numeric = df[['Income','Months Since Policy Inception','Number of Policies']]
```

```python

sns.boxplot(y="Income", data = df_numeric)
```

```python
sns.boxplot(y="Months Since Policy Inception", data = df_numeric)
```

```python
sns.boxplot(y='Number of Policies', data = df_numeric)
```

```python
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
Y = df_numeric['Income']
X = df_numeric.drop(['Income'], axis=1)
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
print(model.summary())
```
